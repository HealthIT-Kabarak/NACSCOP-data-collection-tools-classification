```python
def get_gender(row):
    if 'Female' in row['variable'] or '(F)' in row['variable']:
        return 'Female'
    elif 'Male' in row['variable'] or '(M)' in row['variable']:
        return 'Male'
    else:
        return 'undefined'
def get_type(row):
    if 'Currently on ART' in row['variable']:
        return 'Old'
    elif 'MOH 731   On ART' in row['variable']:
        return 'New'
    else:
        return 'Undefined'
def get_age(row):
    if 'below 1 year' in  row['variable'] or '<1' in  row['variable']:
        return '<1'
    elif 'above 15 years' in row['variable']:
        return '>15'
    elif 'Above 15 years' in row['variable']:
        return '>15'
    elif 'Below 15 years' in row['variable']:
        return '<15'
    elif 'below 15 years' in row['variable']:
        return '<15'
    elif '1-9' in row['variable']:
        return '<10'
    elif '10-14' in row['variable']:
        return '<15'
    elif '15-19' in row['variable']:
        return '<20'
    elif '20-24' in row['variable']:
        return '<25'
    elif '25+' in row['variable']:
        return '>25'
    else:
        return 'Undefined'
def process_row(row):
    gender = get_gender(row)
    type = get_type(row)
    age= get_age(row)
    return pd.Series({'gender': gender, 'age': age,'type': type})
unpivot['variable'] = unpivot['variable'].str.strip()  # remove extra spaces in data column
unpivot[['gender','age', 'type']] = unpivot.apply(process_row, axis=1)  # create a new column 'gender'

# Now the 'gender' column will contain the gender based on the data in each row
unpivot.head(10)
```

```python
#train test split
X=unpivot[['gender','age','value']]
y=unpivot['type']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
# Create a decision tree classifier and fit it to the training data
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = dt_classifier.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# model with gradient boosting
from sklearn.ensemble import GradientBoostingRegressor
g_regressor = GradientBoostingRegressor(random_state=103, n_estimators=200, learning_rate=0.1, max_depth=3, loss='ls', min_samples_split=2, min_samples_leaf=1, subsample=1)
g_regressor.fit(X_train, y_train)

#predicting the test set results
y_pred = g_regressor.predict(X_test)

#evaluating the algorithm
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))

# Calculate accuracy score
regressor_score = g_regressor.score(X_test, y_test)
print("Accuracy score: ", regressor_score)
```
